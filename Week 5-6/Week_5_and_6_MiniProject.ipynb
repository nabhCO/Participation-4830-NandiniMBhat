b) The system of equations below are logistic equations, used for calculating the change in the rabbit population measured
in hectorabbits (x') and the fox population (y'). This results in a representation of the bounded growth of both populations. The model calculates the rate in change of the rabbit and fox populations- it takes birth rate, carrying capacity, natural death and the effects of predation into account to do so, for both species. May's assumptions are incorporated into the system through the subtraction of cxy/x + d (or D(x) * y) in the formula for x', as his assumptions regarding rabbit-fox interactions directly affect the amount by which the rabbit population will change. This is because D(x) * y represents the total number of rabbits eaten over a period of time- this lowers the change in the rabbit population.

```python
#for parts c, d, e, f

import numpy as np
from matplotlib import pyplot as plt

def d_of_x(c, d, x, y):
    
    return (c * x * y) / (x + d)

def RF_PLOT(num_rabbits, num_foxes, t_final, a, b, c, d, e, f ):
    
    t_initial = 0
    t = t_initial
    R = num_rabbits
    F = num_foxes

    # Creating empty arrays to hold values we will plot
    R_graph = []
    F_graph = []
    
    t_vals = np.linspace(0, t_final, 1000)
    delta_t = (t_final - t_initial) / 1000
    
    for k in t_vals:
        
        R_graph.append(R)
        F_graph.append(F)
        
        Rprime = (a*R) * (1 - (R/b)) - d_of_x(c, d, R, F)
        Fprime = (e*F) * (1 - (F/(f*R)))
        
        delta_R = Rprime*delta_t
        delta_F = Fprime*delta_t
        
        R = R + delta_R
        F = F + delta_F
        
    plt.plot(t_vals, R_graph, label= "Rabbit Population")
    plt.plot(t_vals, F_graph, label= "Fox Population")
    plt.xlabel("Time (months)")
    plt.ylabel("Population")
    plt.title("Population Change Over Time")
    plt.legend()
    plt.show()
    
RF_PLOT(20, 10, 100, 0.6, 10, 0.5, 1, 0.1, 2)
        
```

### Q2- Fermentation

8a) A logistic equation must have a growth rate and carrying capacity for the variable it measures. If the growth rate is a, the carrying capacity is b, and the current amount of yeast is Y, we can form the equation aY * (1 - (Y/b)) to describe the bounded growth of the yeast. This can be simplified to aY - (a * (Y ** 2)) / b. Therefore, Y(t) = aY - (a * (Y ** 2)) / b.

```python
#for parts b and c

def growth_func(a, yeast, b):
    
    return (a * yeast) - (a * (yeast**2)) / b

def Y_PLOT(weight_yeast, t_final):
    
    t_initial = 0
    t = t_initial
    Y = weight_yeast
    Y_half = weight_yeast

    # Creating empty arrays to hold values we will plot
    Y_graph = []
    Y_half_graph = []
    
    t_vals = np.arange(0, t_final, 0.001)
    delta_t = 0.001
    
    for k in t_vals:
        
        Y_graph.append(Y)
        Y_half_graph.append(Y_half)
        
        Yprime = growth_func(0.2, Y, 10)
        Y_half_prime = growth_func(0.1, Y_half, 10)
       
        delta_Y = Yprime * delta_t
        delta_Y_half = Y_half_prime * delta_t
        
        Y = Y + delta_Y
        Y_half = Y_half + delta_Y_half
        
    Y_graph = np.array(Y_graph)
    Y_half_graph = np.array(Y_half_graph)
    
    half_cap = (np.abs(Y_graph - 5.0)).argmin()   #technique I found on stackoverflow for locating specific points (or closest to specific point) in a numpy array, link in resources
    onep_cap = (np.abs(Y_graph - 9.9)).argmin()
    half_cap_2 = (np.abs(Y_half_graph - 5.0)).argmin()
    onep_cap_2 = (np.abs(Y_half_graph - 9.9)).argmin()
    
    plt.plot(t_vals, Y_graph, label = "Growth Rate 0.2")
    plt.plot(t_vals[half_cap], Y_graph[half_cap], marker = "x", label = "Half Capacity")
    plt.plot(t_vals[onep_cap], Y_graph[onep_cap], marker = "x", label = "One percent")
    
    plt.plot(t_vals, Y_half_graph, label = "Growth Rate 0.1")
    plt.plot(t_vals[half_cap_2], Y_half_graph[half_cap_2], marker = "x", label = "Half Capacity")
    plt.plot(t_vals[onep_cap_2], Y_half_graph[onep_cap_2], marker = "x", label = "One percent")
    
    plt.xlabel("Time (hours)")
    plt.ylabel("Yeast (lbs)")
    plt.title("Yeast Growth")
    plt.legend()
    plt.grid(True)
    
    
    plt.show()
   
Y_PLOT(0.5, 100)
        
```

9a) From the description given in the question, it is clear that the rate of alcohol production depends on the current amount of yeast. Since each pound of yeast produces 0.05 lbs of alcohol per hour, we write the equation for alcohol production as A(t) = 0.05Y.

b) Alcohol killing the yeast means that it plays a role in reducing the amount of yeast produced at each time step. This will result in a term being subtracted from the equation for production. We have to consider that in each pound of yeast, one pound of alcohol will kill 0.1 lbs of yeast per hour. Following the way the Lotka-Volterra model works, we can simply put this as the alcohol-yeast interactions (if we are looking at the yeast as a "predator"). This gives us the term 0.1YA. The equation for yeast becomes aY - ((a * (Y ** 2)) / b) - 0.1YA.

```python
#for part c

def growth_func(a, yeast, b, alcohol, c):
    
    return (a * yeast) - ((a * (yeast**2)) / b) - (c * yeast * alcohol)

def YA_PLOT(weight_yeast, weight_alc, t_final, a, b, c, alc_prod):
    
    t_initial = 0
    t = t_initial
    Y = weight_yeast
    A = weight_alc

    # Creating empty arrays to hold values we will plot
    Y_graph = []
    A_graph = []
    
    t_vals = np.arange(0, t_final, 0.001)
    delta_t = 0.001
    
    for k in t_vals:
        
        Y_graph.append(Y)
        A_graph.append(A)
        
        Yprime = growth_func(a, Y, b, A, c)
        Aprime = alc_prod * Y
       
        delta_Y = Yprime * delta_t
        delta_A = Aprime * delta_t
        
        Y = Y + delta_Y
        A = A + delta_A
        
    Y_graph = np.array(Y_graph) 
    low_yeast = (np.abs(Y_graph - 0.01)).argmin()
    
    plt.plot(t_vals, Y_graph, label = "Yeast")
    plt.plot(t_vals, A_graph, label = "Alcohol")
    plt.plot(t_vals[low_yeast], Y_graph[low_yeast], marker = "x", label = "0.1 lbs")
    
    plt.xlabel("Time (hours)")
    plt.ylabel("Weight (lbs)")
    plt.title("Yeast and Alcohol Growth")
    plt.legend()
    plt.grid(True)
    
    
    plt.show()
   
YA_PLOT(0.5, 0, 100, 0.2, 10, 0.1, 0.05)
        
```

```python
#for part c

def YAS_PLOT(weight_yeast, weight_alc, weight_sugar, t_final):
    
    #initial values
    t_initial = 0
    t = t_initial
    Y = weight_yeast
    A = weight_alc
    S = weight_sugar

    Y_graph = []
    A_graph = []
    S_graph = []
    
    t_vals = np.arange(0, t_final, 0.001)
    delta_t = 0.001
    
    
    for k in t_vals:

        Y_graph.append(Y)
        A_graph.append(A)
        S_graph.append(S)
        
        #equations representing growth rate of yeast, alcohol and sugar
        Yprime = (0.2 * Y) - ((0.2 * (Y**2)) / (0.4 * S)) - (0.1 * Y * A)
        Aprime = 0.05 * Y
        Sprime = -0.15 * Y
       
        delta_Y = Yprime * delta_t
        delta_A = Aprime * delta_t
        delta_S = Sprime * delta_t
        
        Y = Y + delta_Y
        A = A + delta_A
        S = S + delta_S
        
    Y_graph = np.array(Y_graph) 
    low_yeast = (np.abs(Y_graph - 0.01)).argmin()
        
    
    plt.plot(t_vals, Y_graph, label = "Yeast")
    plt.plot(t_vals, A_graph, label = "Alcohol")
    plt.plot(t_vals, S_graph, label = "Sugar")
    plt.plot(t_vals[low_yeast], Y_graph[low_yeast], marker = "x", label = "0.1 lbs")
    
    plt.xlabel("Time (hours)")
    plt.ylabel("Growth of Yeast, Alcohol, Sugar")
    plt.title("Weight (lbs)")
    plt.legend()
    plt.grid(True)
    
    plt.show()
   
YAS_PLOT(0.5, 0, 25, 100)
```

### Q3- Leaky Integrate and Fire Model

For this question, I decided to try and follow an assignment from MIT's "Brains, Minds, and Machines" summer course (available on MIT Opencourseware- I have added the link to the assignment in the references below). The assignment involves implementing the "Leaky Integrate-and-Fire" model in MATLAB, which is a fundamental model describing the electrical response of a neuron to an input (which is a current injected into the cell). The model is basic and doesn't describe many aspects of how a neuron works to propagate signals, but it is the precursor to more advanced models such as the Hodgkin-Huxley model. It can also be tweaked to increase its biological accuracy. I thought it would be interesting to try this out and play around with the model a bit, in Python instead of MATLAB.

I will begin with a simple explanation of the model. When these is no current injected, and no signals being progagated, the neuron is at its resting potential. The resting potential is the difference between the charges on the inside and outside of a neuron's membrane. When a current is injected into the neuron, its membrane potential will rise at a rate dependent on the injected current. If the membrane potential reaches a certain threshold, this causes a "spike", or action potential (this means that the neuron has fired). After this spike, the membrane potential "resets" to the resting potential. After this, the membrane potential displays a refractory period- during this time, it cannot spike again. This is due to the mechanisms of the membrane's ion channels. This pattern of the membrane potential increasing, spiking, resetting and going through the refractory period continues as long as adequate current is applied.

The parameters of the model start with the neuron's capacitance Cm (its ability to store charge) and its resistance Rm (ability to oppose a current). Both of these allow us to calculate the membrane time constant, represented by Tau_m in the code. The membrane time constant tells us how quickly the membrane takes to respond to the applied stimulus, which is the current in this case. The other parameters include the resting potential El, the threshold Vth, the injected current Ie, and the duration over which the membrane potential is to be calculated. The differential equation that describes the change in membrane potential over time is (El - V) + (RmIe(t))) / Tau_m. Ie(t) would represent a change in injected current over time, if there were changes made to it. This equation comes from the fact that the model is based on a resistor-capacitor circuit- the "spiking" of the neuron, as well as the refractory period, are added in to make the model more biologically accurate.

I calculated the appropriate count with 3/0.01, which is 300. As a note, the resistor-capacitory model mentioned earlier does not graph spikes, so they are added in at the points where a spike occurs. These are the vertical dotted lines on the graph.

```python
def LIF_Model(capacitance, resistance, resting_potential, threshold, injected_current, delta_t, t_final):
    
    Cm = capacitance
    Rm = resistance
    El = resting_potential
    Vth = threshold
    Ie = injected_current
    Tau_m = Cm * Rm
    
    
    V_vals = []
    spike_vals = []
    spike_times = []
    t_vals = np.arange(0, t_final, delta_t)
    V = El
    
    refrac_period = False
    refrac_count = 0
    
    for t in t_vals:
        
        V_vals.append(V)
        
        if refrac_period == True:
            
            if refrac_count < 300:
                refrac_count += 1
                spike_vals.append(0)
                continue
                
            else:
                refrac_period = False
                refrac_count = 0
            
            
        Vprime = ((El - V) + (Rm * Ie)) / Tau_m
        delta_V = Vprime * delta_t
        V = V + delta_V
        
        if V >= Vth:
            
            spike_vals.append(1)
            spike_times.append(t)
            V = El
            refrac_period = True
            
            
        else:
            spike_vals.append(0)
            
               
    plt.plot(t_vals, V_vals, label = "Voltage")
    
    for i in range(len(spike_vals)):
        
        if spike_vals[i] == 1:
            plt.axvline(x = t_vals[i], color = "b", linestyle = "--", linewidth = 1)
    
    
        
    plt.xlabel("Time (msec)")
    plt.ylabel("Membrane Potential (mV)")
    plt.title("Leaky Integrate-and-Fire")
    plt.legend()
    plt.grid(True)
    
    plt.show()
    
    
LIF_Model(200, 100, -70, -60, 150, 0.01, 150)
```

The ability of Rm to change spike intervals may depend on interactions with a different constant, such as the injected current or Cm itself.
